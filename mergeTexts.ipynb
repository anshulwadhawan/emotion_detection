```
import string
import re
import operator
import os 

for file_name in os.listdir():

  fp = open(file_name, 'r')

  current_id=''
  for line in fp:
      line = line.strip()
      if len(line)>0:
        if line.isdigit():
          #print(line)
          current_id = line
        else:
          #print("hi")
          #print(line)
          id_tweet_map[current_id]=line  
```

```
import string
import re
import operator
import os 

ids=[]
for file_name in os.listdir():

  fp = open(file_name, 'r')

  for line in fp:
      line = line.strip()
      if len(line)>0:
        if line.isdigit():
          #print(line)
          ids.append(line)
```
